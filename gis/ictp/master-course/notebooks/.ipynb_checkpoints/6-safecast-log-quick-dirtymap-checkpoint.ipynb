```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
import numpy as np
from palettable.colorbrewer.sequential import YlOrRd_5

%matplotlib inline

import pandas as pd
import geopandas as gpd
from shapely.geometry import Point
```

```python
f = 'data/safecast_subset.csv'    # Path to file containing the table
df = pd.read_csv(f)    # Read the table in

# Convert to geo-dataframe
df['geometry'] = df.apply(lambda z: Point(z.latitude, z.longitude), axis=1)
df = gpd.GeoDataFrame(df)

df.crs = {'init' :'epsg:4326'}

plt.rcParams["figure.figsize"] = [15, 15]

bounds = df.total_bounds

map = Basemap(llcrnrlon=bounds[1],llcrnrlat=bounds[0],
              urcrnrlon=bounds[3],urcrnrlat=bounds[2], 
              epsg=4326)

map.hexbin(df.longitude.values, df.latitude.values, C = df.value.values, 
           reduce_C_function = max, gridsize=100, 
           mincnt=1, cmap=YlOrRd_5.mpl_colormap, 
           linewidths=0.5, edgecolors='none')

map.colorbar(location='bottom', label='Counts per minute (CPM)')
map.arcgisimage(service='ESRI_Imagery_World_2D', dpi=300, verbose= True)

plt.show()
```
